## Função de Consulta
A função abaixo executa a consulta, exibe o retorno da consulta no banco e grava os dados necessários no arquivo resultado.

```python
#NÃO ALTERE AS LINHAS ABAIXO
def runQuery(query, query_name):
  df = pd.read_sql(query, con)
  query_size = df.shape[0]
  writer.writerow([query_name, query_size, query])
  display(df)
```

## PARTE 1 - Transformar Algebra Relacional para SQL
Abaixo, seguem as consultas que voce terá de transformar de Algebra Relacional para SQL. Altere **APENAS** o espaço entre as aspas. Dentro desse espaço é que você deverá digitar o comando de execução da consulta.

### 1 - $\pi_{id\_viagem,origem\_cidade,destino\_cidade} \: \sigma_{transporte='Rodoviario'}$ (trecho)


```python
a1 = "SELECT id_viagem, origem_cidade, destino_cidade FROM trecho WHERE (transporte = 'Rodoviário')"
#NÃO ALTERE A LINHA ABAIXO
runQuery(a1, "a1")
```

```python
### 2 - $\pi_{id\_pagamento,valor} \: \sigma_{valor>1000}$ (pagamento)
```

```python
a2 = "SELECT id_pagamento, valor FROM pagamento WHERE (valor > 1000)"
#NÃO ALTERE A LINHA ABAIXO
runQuery(a2, "a2")
```

### 3 - $\pi_{orgao.nome\_orgao,pagamento.valor} \: \sigma_{pagamento.id\_orgao\_superior=orgao.id\_orgao \wedge pagamento.valor < 200}$ (pagamento $\times$ orgao)

```python
a3 = "SELECT orgao.nome_orgao, pagamento.valor FROM pagamento FULL OUTER JOIN orgao WHERE(pagamento.id_orgao_superior = orgao.id_orgao AND pagamento.valor < 200)"
#NÃO ALTERE A LINHA ABAIXO
runQuery(a3, "a3")
```

### 4 - $\pi_{id\_viagem,sequencia\_trecho,destino\_viagem} \: \sigma_{sequencia\_trecho=5}$ (viagem $\bowtie$ trecho)

```python
a4 = "SELECT viagem.id_viagem, trecho.sequencia_trecho, viagem.destino_viagem FROM viagem JOIN trecho ON viagem.id_viagem = trecho.id_viagem WHERE(trecho.sequencia_trecho = 5)"
#NÃO ALTERE A LINHA ABAIXO
runQuery(a4, "a4")
```

### 5 - $\pi_{id\_pagamento,valor} \: \sigma_{viagem\_urgente=SIM}$ (pagamento $\bowtie$ viagem)


```python
a5 = "SELECT id_pagamento, valor FROM pagamento JOIN viagem ON viagem.id_viagem = pagamento.id_viagem WHERE(viagem_urgente = 'SIM')"
#NÃO ALTERE A LINHA ABAIXO
runQuery(a5, "a5")
```

## PARTE 2 - Consultas SQL
Abaixo, seguem as consultas que voce terá de executar. Altere **APENAS** o espaço entre as aspas. Dentro desse espaço é que você deverá digitar o comando de execução da consulta.

### 1 - Liste os campos id\_viagem, situacao, nome\_viajante, data\_inicio, data\_fim e destino\_viagem para todas as viagens com data de inicio entre 2022-01-09 e 2022-01-16 (incluso):

**OBS:** Se atentar para o formato da data no banco (ano-mês-dia)

```python
q1 = "SELECT id_viagem, situacao, nome_viajante, data_inicio, data_fim, destino_viagem FROM viagem WHERE data_inicio BETWEEN '2022-01-09' AND '2022-01-16'"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q1, "q1")
```

### 2 - Liste todos os códigos e nomes dos órgãos que tenham a palavra "Nacional" no nome:

```python
q2 = "SELECT codigo_orgao, nome_orgao FROM orgao WHERE nome_orgao LIKE '%Nacional%'"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q2, "q2")
```

### 3 - Liste os campos nome\_viajante e cargo\_funcao\_viajante dos servidores que tenham nome que termina com "SANTOS"
**Obs:** pesquisar a palavra em caixa alta

```python
q3 = "SELECT nome_viajante, cargo_funcao_viajante FROM viagem WHERE UPPER(nome_viajante) LIKE '%SANTOS'"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q3, "q3")
```

### 4 - Liste os 50 primeiros nomes dos servidores e valores das passagens que tiveram "São Paulo" como o estado de origem da ida da passagem. O resultado deverá ser ordenado primeiramente pelo valor e depois pelo nome, os dois em ordem crescente:

```python
q4 = "SELECT nome_viajante, valor_passagem FROM viagem v JOIN passagem p ON v.id_viagem = p.id_viagem WHERE p.ida_uf_origem = 'São Paulo' ORDER BY valor_passagem, nome_viajante LIMIT 50"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q4, "q4")
```

### 5 - Liste o nome do órgão superior e a quantidade de viagens urgentes realizadas que cada órgão superior pagou. Em orderm crescente de número de viagens:

```python
q5 = "SELECT orgao.nome_orgao AS orgao_superior, COUNT(viagem.id_viagem) AS quantidade_viagens_urgentes FROM pagamento JOIN viagem ON pagamento.id_viagem = viagem.id_viagem JOIN orgao ON pagamento.id_orgao_superior = orgao.id_orgao WHERE viagem.viagem_urgente = 'SIM' GROUP BY orgao.nome_orgao ORDER BY quantidade_viagens_urgentes ASC"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q5, "q5")
```

### 6 - Liste o campo id\_viagem e o total do campo valor do pagamento de viagens solicitadas pelo órgão de código 52000:

```python
q6 = "SELECT viagem.id_viagem, SUM(pagamento.valor) as total_pagamento FROM viagem JOIN pagamento ON viagem.id_viagem = pagamento.id_viagem JOIN orgao ON viagem.id_orgao_solicitante = orgao.id_orgao WHERE orgao.codigo_orgao = 52000 GROUP BY viagem.id_viagem"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q6, "q6")
```

### 7 - Liste os ids das viagens que tenham mais de 3 passagens (4 ou mais) e o valor da passagem mais barata de cada uma:

```python
q7 = "SELECT v.id_viagem, MIN(p.valor_passagem) AS valor_passagem_mais_barata FROM viagem v JOIN passagem p ON v.id_viagem = p.id_viagem GROUP BY v.id_viagem HAVING COUNT(p.id_passagem) > 3"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q7, "q7")
```

### 8 - Liste todos os ids dos trechos das viagens que não tem "Brasil" como país de origem e que o viajante possui a palavra "MINISTRO" no nome da função/cargo:

```python
q8 = "SELECT trecho.id_trecho FROM trecho JOIN viagem ON trecho.id_viagem = viagem.id_viagem WHERE trecho.origem_pais != 'Brasil' AND viagem.cargo_funcao_viajante LIKE '%MINISTRO%' ORDER BY trecho.id_trecho"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q8, "q8")
```

### 9 - Liste, em ordem alfabética, os nomes de todos os órgãos superiores que custearam as despesas das viagens que tiveram gastos com diárias maiores que 0 e menores que 1000.

```python
q9 = "SELECT DISTINCT o.nome_orgao FROM orgao o JOIN viagem v ON o.id_orgao = v.id_orgao_superior WHERE v.valor_diarias > 0 AND v.valor_diarias < 1000 ORDER BY o.nome_orgao"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q9, "q9")
```

### 10 - Liste os ids das viagens urgentes que têm algum trecho em que o transporte não foi "Aéreo" nem "Rodoviário", o nome do viajante e o valor total dos gastos da viagem indicados na tabela pagamento (com o nome "valor\_total\_pago\_viagem"). O resultado deverá considerar apenas as viagens que tiveram gastos com "PASSAGEM" e deverá ser exibido em ordem alfabética em relação ao nome do viajante:

```python
q10 = "SELECT viagem.id_viagem, viagem.nome_viajante, (viagem.valor_diarias + viagem.valor_passagens + viagem.valor_devolucao + viagem.valor_outros_gastos) AS valor_total_pago_viagem FROM viagem JOIN trecho ON viagem.id_viagem = trecho.id_viagem WHERE trecho.transporte NOT IN ('Aéreo', 'Rodoviário') AND viagem.viagem_urgente = 'SIM' AND viagem.valor_diarias = 0 AND viagem.valor_devolucao = 0 AND viagem.valor_outros_gastos = 0 AND viagem.valor_passagens != 0 ORDER BY viagem.nome_viajante"
#NÃO ALTERE A LINHA ABAIXO
runQuery(q10, "q10")
```
